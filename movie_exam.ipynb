# Question 1
Who won the Oscar for “Actor in a Leading Role” in 2015?

(Hint: The winner is indicated as '1.0'.)


```python
%%sql

SELECT 
    *
FROM 
    oscars
WHERE    
    year = 2015
AND    
    award = 'Actor in a Leading Role'
AND 
    winner = 1.0;
    
```

# Question 2
What query will produce the ten oldest movies in the database?



```python
%%sql

SELECT 
    * 
FROM 
    movies 
WHERE 
    release_date IS NOT NULL 
ORDER BY 
    release_date ASC 
LIMIT 10;

```

# Question 5
How many movies are there that are both in the "Thriller" genre and contain the word “love” anywhere in the keywords?


```python
%%sql

SELECT 
    COUNT(distinct m.movie_id) AS num_of_movies
FROM 
    movies AS m
INNER JOIN
    genremap AS g
ON    
    m.movie_id = g.movie_id
INNER JOIN     
    keywordmap AS k
ON     
    m.movie_id = k.movie_id
WHERE  
    keyword_id in (SELECT 
                      keyword_id
                   FROM 
                      keywords
                   WHERE 
                      keyword_name LIKE '%love%'
                   )  
AND 
    genre_id in (SELECT 
                    genre_id
                FROM 
                    genres
                WHERE 
                    genre_name = 'Thriller'
                );    

    
```

# Question 6
How many movies are there that were released between 1 August 2006 ('2006-08-01') and 1 October 2009 ('2009-10-01') that have a popularity score of more than 40 and a budget of less than 50 000 000?



```python
%%sql

SELECT 
    COUNT(*) AS num_of_movies
FROM 
    movies
WHERE 
    release_date between '2006-08-01' AND '2009-10-01'
AND 
    popularity > 40
AND    
    budget < 50000000;
    
```

# Question 7
How many unique characters has "Vin Diesel" played so far in the database?



```python
%%sql

SELECT 
    count(distinct c.characters) AS num_of_Vin_Diesel_characters
FROM
    movies AS m
INNER JOIN 
    casts AS c
ON 
    m.movie_id = c.movie_id
INNER JOIN 
    actors AS a
ON 
    a.actor_id = c.actor_id
WHERE 
    actor_name = 'Vin Diesel';
```

# Question 8
What are the genres of the movie “The Royal Tenenbaums”?



```python
%%sql

SELECT
    genre_name
FROM    
    genres AS gs
INNER JOIN     
    genremap AS g
ON    
    g.genre_id = gs.genre_id
WHERE     
    movie_id = (SELECT 
                    movie_id
                FROM
                    movies
                WHERE
                    title = 'The Royal Tenenbaums')
    ;
    
```

# Question 9
What are the three production companies that have the highest movie popularity score on average, as recorded within the database?



```python
%%sql

SELECT 
    production_company_name, 
    round(AVG(popularity),2) AS avg_popularity   
FROM    
    productioncompanies AS p
INNER JOIN    
    productioncompanymap AS pp
ON    
    p.production_company_id = pp.production_company_id
INNER JOIN
    movies AS m
ON    
    m.movie_id = pp.movie_id
GROUP BY 
    production_company_name
ORDER BY   
    AVG(popularity) DESC 
LIMIT 3;
```

# Question 10

# Question 11¶
Which genre has, on average, the lowest movie popularity score?



```python
%%sql

SELECT 
    g.genre_name, 
    round(AVG(m.popularity),2) AS AVG_popularity
FROM    
    movies AS m
INNER JOIN    
    genremap AS gp
ON    
    m.movie_id = gp.movie_id
INNER JOIN    
    genres AS g
ON    
    g.genre_id = gp.genre_id
GROUP BY
    g.genre_name
ORDER BY 
    AVG_popularity 
LIMIT 1;
```

# Question 12
Which award category has the highest number of actor nominations (actors can be male or female)? (Hint: Oscars.name contains both actors' names and film names.)



```python
%%sql

SELECT 
    award,
    COUNT(*) AS num_of_awards
FROM
    oscars
WHERE 
    winner IS NULL 
AND 
    name IN (SELECT 
                 DISTINCT actor_name
             FROM
                 actors)
GROUP BY    
    award
ORDER BY  
    num_of_awards DESC 
LIMIT 1;
```

# Question 14
DStv will be having a special week dedicated to the actor Alan Rickman. Which of the following queries would create a new view that shows the titles, release dates, taglines, and overviews of all movies that Alan Rickman has played in?



```python
%%sql

DROP VIEW IF EXISTS Alan_Rickman_Movies;

CREATE VIEW Alan_Rickman_Movies AS
    SELECT 
        title, 
        release_date, tagline, 
        overview
    FROM 
        Movies
    LEFT JOIN 
        Casts
    ON 
        Casts.movie_id = Movies.movie_id 
    Left JOIN 
        Actors 
    ON 
        Casts.actor_id = Actors.actor_id 
    WHERE 
        Actors.actor_name = 'Alan Rickman';
        
SELECT 
    *
FROM
    Alan_Rickman_Movies;
```
